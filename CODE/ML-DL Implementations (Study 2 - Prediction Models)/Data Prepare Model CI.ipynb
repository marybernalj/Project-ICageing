Rename file columns and add the year

```python
import pandas as pd

def renombrar_y_agregar_ano(archivo_entrada, archivo_salida, nuevos_nombres, ano):
    """
    Renombra todas las columnas de un archivo CSV y agrega una columna con el año.

    Args:
        archivo_entrada (str): Ruta del archivo CSV de entrada.
        archivo_salida (str): Ruta del archivo CSV de salida.
        nuevos_nombres (list): Lista de nuevos nombres para las columnas.
        ano (int): Año a agregar.
    """

    # Leemos el archivo CSV
    df = pd.read_csv(archivo_entrada)

    # Renombramos todas las columnas a la vez
    df.columns = nuevos_nombres

    # Agregamos la columna 'año' con el valor especificado
    df['año'] = ano

    # Guardamos el DataFrame modificado en un nuevo archivo CSV
    df.to_csv(archivo_salida, index=False)

# Ejemplo de uso:
archivo_entrada = "/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/H06T_R.csv"
archivo_salida = "/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2006.csv"
nuevos_nombres = ["ID", "AGE", "SEX", "ALONE", "LOCOMOTION", "SENSORY", "VITALITY", "PSICHOLOGICAL", "COGNITION", "PHYSICAL_ACTIVITY", "SMOKE", "CIGARRETTES", "TOMA", "DRINK", "CALC_IMC", "IMC", "NURSING_HOME" ]
ano = 2006

renombrar_y_agregar_ano(archivo_entrada, archivo_salida, nuevos_nombres, ano)
```

FULL FILE UNION ONE RECORD BELOW THE OTHER ORDERED BY ID AND YEAR

```python
import pandas as pd
import numpy as np

def unir_archivos(lista_archivos):
    """
    Une múltiples archivos CSV en un solo DataFrame, apilando las filas.

    Args:
        lista_archivos: Una lista con las rutas de los archivos CSV.

    Returns:
        Un DataFrame de Pandas con todos los datos combinados.
    """

    df_final = pd.DataFrame()
    for archivo in lista_archivos:
        df_temp = pd.read_csv(archivo)
        df_final = pd.concat([df_final, df_temp], ignore_index=True)
    return df_final


lista_archivos = ['/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2008.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2010.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2012.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2014.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2016.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2018.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2020.csv', '/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/2022.csv']
df_completo = unir_archivos(lista_archivos)
df_ordenado = df_completo.sort_values(by=['ID', 'año'])
df_ordenado.to_csv('/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/completo_ordenado.csv', index=False)
```

ADD SEQUENTIAL VALUATION COLUMN

```python
import pandas as pd

def agregar_columna_evaluacion(df):
    """
    Agrega una columna de evaluación secuencial para cada combinación única de ID y año.

    Args:
        df (pd.DataFrame): DataFrame de entrada.

    Returns:
        pd.DataFrame: DataFrame con la columna de evaluación agregada.
    """

    # Agrupar por ID y año, y numerar cada grupo
    df['evaluacion'] = df.groupby(['ID']).cumcount() + 1
    return df

# Leer el archivo CSV
df = pd.read_csv('/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/completo_ordenado.csv')

# Ordenar por ID y año (si no está ordenado previamente)
df = df.sort_values(by=['ID', 'año'])

# Agregar la columna de evaluación
df = agregar_columna_evaluacion(df)

# Guardar el DataFrame con la nueva columna
df.to_csv('/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/DATOS_OBSERVACION.csv', index=False)
```

PREPROCESSING: TRANSFORMING FILE VARIABLES

```python
import pandas as pd
import numpy as np

def cambiar_estado_civil(df, columna):
   df[columna] = df[columna].replace({
      1: 'PARTNERED',
      2: 'PARTNERED',
      3: 'PARTNERED',
      4: 'PARTNERED',
      6: 'UNCOUPLED'
   })
   return df

def cambiar_physical_activity(df, columna):
  df[columna] = df[columna].replace({
      1: 'SEDENTARY',
      2: 'MILD',
      3: 'MODERATE',
      4: 'VIGOROUS'
  })
  return df

def cambiar_gender(df, columna):
  df[columna] = df[columna].replace({
      1: 'MALE',
      2: 'FEMALE'
  })
  return df

# Leer el archivo CSV
df = pd.read_csv('/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/DATOS_OBSERVACION_VAR_NOTNULL2010-2022.csv')
df = cambiar_estado_civil(df, 'ALONE')
df = cambiar_physical_activity(df, 'PHYSICAL_ACTIVITY')
df = cambiar_gender(df, 'SEX')
df['CIGARRETTES'] = df['CIGARRETTES'].fillna(0)
df['DRINK'] = df['DRINK'].fillna(0)
new_df=df[['ID','AGE','SEX','ALONE','LOCOMOTION','SENSORY','VITALITY','PSICHOLOGICAL','COGNITION','PHYSICAL_ACTIVITY','CIGARRETTES','DRINK','CALC_IMC','año','evaluacion']]
# Guardar el DataFrame con la nueva columna
new_df.to_csv('/content/drive/My Drive/PREDICTIVE MODEL IC/DATA/DATOS_OBSERVACION_VAR_NOTNULL2010-2022_TRANSFORM.csv', index=False)
new_df.shape
```
